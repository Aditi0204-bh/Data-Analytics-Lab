# factorial of a number

```python
#factorial of a number
def factorial(q):
    if q==1:
        return 1
    else:
        return (q * factorial(q-1))

p=input("Enter the number")
q=int(p)
z = factorial(q)
print("factorial is: ",z)
```

```python
def factorial(num):
    if num==0:
        print("factorial is 1")
    else:
        for i in range(1,num+1):
            num=num*i
        print(num)

a =input("Enter the number whoes factorial is to be calculated")
num =int(a)
factorial(num)
```
